```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:\\Users\\gymin\\OneDrive\\Documents\\CPIchangesforMacro.xlsx")
```

```python
# Convert the "Date" column to DateTime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the "Date" column
df['Year'] = df['Date'].dt.year

# Group data by year and calculate the mean of CPI changes
yearly_data = df.groupby('Year')['CPIchanges'].mean()

# Separate positive and negative CPI changes
positive_changes = yearly_data[yearly_data >= 0]
negative_changes = yearly_data[yearly_data < 0]

# Create the bar plot
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.bar(positive_changes.index, positive_changes, color='green', label='Positive CPI Changes')
plt.bar(negative_changes.index, negative_changes, color='red', label='Negative CPI Changes')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Average CPI Changes')
plt.title('Average Positive and Negative CPI Changes by Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
